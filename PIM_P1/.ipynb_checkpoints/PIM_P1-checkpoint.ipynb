### Ejercicio 1

```python
import numpy as np
from skimage import io, transform
from matplotlib import pyplot as plt
```

Crearemos una imagen hibrida entre la imagen de un gato y una persona usando convoluciones en los 3 canales de cada imagen. Para esto necesitamos que las dos imagenes sean del mismo tamaño y esten centrada para que el resultado sea correcto. El siguiente codigo convierte las dos imagenes al mismo tamaño y recorta la imagen de la persona para cuadrarla con la del gato.

```python
cat = io.imread('cat.png').astype(np.uint8)
human = io.imread('human.png').astype(np.uint8)

size1 = int(round(human.shape[0]*(4/3.)));
size2 = int(round(cat.shape[1]*(4/3.)));
human_resize = transform.resize(human, (size1, size2));
left = 50
top = 50
right = 51+cat.shape[0]-1
bottom = 51+cat.shape[1]-1 
human_crop = human_resize[left: right, top: bottom]

fig, axarr = plt.subplots(ncols=2, nrows=1,figsize=(15,5))
pos1 = axarr[0].imshow(cat)
pos2 = axarr[1].imshow(human_crop)
fig.colorbar(pos1, ax=axarr[0])
fig.colorbar(pos2, ax=axarr[1])
plt.show()
print(cat.shape, human_crop.shape)
```

Creamos una funcion gausiana de un tamaño fijo de 6*sigma+1 (55), este tamaño es el que utilizaremos para el kernel de la convolucion, tambien pasamos un parametro Inv que es para hacer la inversa de la gausiana y mostramos los resultado.

```python
def gaussian(sigma = 9, inv = False):
    y,x = np.ogrid[-(6*sigma+1)/2:(6*sigma+1)/2,-(6*sigma+1)/2:(6*sigma+1)/2]
    if inv: g = 1-np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    else: g = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    return g / g.sum()
 
gaussian_kernel_normal = gaussian(sigma=9)
gaussian_kernel_inv = gaussian(sigma=9, inv=True)

print(gaussian_kernel_normal.shape)
plt.imshow(gaussian_kernel_normal, cmap=plt.get_cmap('jet'))
plt.colorbar()
plt.show()

plt.imshow(gaussian_kernel_inv, cmap=plt.get_cmap('jet'))
plt.colorbar()
plt.show()
```

Procedemos a hacer la convolucion con estos filtros, en este caso solo con el primer filtro gausiano, sin la inversa, ya que por defecto da problemas y es mas sencillo restar la imagen original con la obtenida con el filtro pasa bajos y asi conseguir una con los detalles (filtro pasa alto).

Tambien tenemos una funcion que hace la convolucion con dicho filtro/kernel para cada uno de los canales de la imagen (r,g,b) y nos desvuelve la imagen a color con el filtro aplicado. Tenemos otra funcion para normalizar la imagen entre los valores 0-1, esto nos ayudara a visualizar mejor las imagenes en los plots.

Una vez obtenida las imagen con el filtro pasa bajos y la imagen con el filtro pasa alto (detalles) lo unico que tenemos que hacer es sumar ambas imagenes y normalizar el resultado para obtener una imagen hibrida.

```python
from scipy import ndimage, signal
from skimage import img_as_float, img_as_ubyte, color, filters

def conv_rgb_image(image, mask):
    img_copy = np.array(image)
    img_mat = np.zeros((img_copy.shape[0], img_copy.shape[1], 3))
    for i in range(image.shape[2]):
        img_mat[:,:,i] = ndimage.convolve(image[:,:,i], mask, mode="constant",cval=0.0)
    return img_mat

def normalize_img(image):
    for i in range(image.shape[2]):
        image[:,:,i]=((image[:,:,i]-(np.min(image[:,:,i])))/((np.max(image[:,:,i]))-(np.min(image[:,:,i]))))
    return image


human_lp = conv_rgb_image(human_crop, gaussian_kernel_normal)
cat_lw = conv_rgb_image(img_as_float(cat), gaussian_kernel_normal)
cat_hp = normalize_img(img_as_float(cat) - cat_lw)
hybrid = normalize_img(cat_hp + human_lp)

fig, axarr = plt.subplots(ncols=3, nrows=1,figsize=(15,5))
pos1 = axarr[0].imshow(human_lp)
pos2 = axarr[1].imshow(cat_hp, vmin=0, vmax=1)
pos3 = axarr[2].imshow(hybrid, vmin=0, vmax=1)
fig.colorbar(pos1, ax=axarr[0])
fig.colorbar(pos2, ax=axarr[1])
fig.colorbar(pos3, ax=axarr[2])
plt.show()
```

En esta segunda parte tenemos sustituir la convolucion aplicando la transformada de fourier a la imagen y al filtro para despues multiplicarlos y obtener el mismo resultado. Para esto, tenemos el mismo problema que en el anterior apartado, tenemos imagenes con 3 canales (r,g,b) con lo cual hay que aplicar la transformada para cada canal.

```python
fig, axarr = plt.subplots(ncols=3, nrows=1,figsize=(15,5))
pos1 = axarr[0].imshow(human_crop[:,:,0], cmap='gray')
pos2 = axarr[1].imshow(human_crop[:,:,1], cmap='gray')
pos3 = axarr[2].imshow(human_crop[:,:,2], cmap='gray')
fig.colorbar(pos1, ax=axarr[0])
fig.colorbar(pos2, ax=axarr[1])
fig.colorbar(pos3, ax=axarr[2])
plt.show()
```

Con lo cual he creado una funcion gausiana con el tamaño para estas imagenes 512, ya que por defecto la funcion de fourier (fft) necesita imagenes con un tamaño de 2^n, con lo cual para estas imagenes necesitamos aplicar un padding para lograr esos 512. La funcion gausiana para esta parte crea una gausiana de dicho tamaña y la funcion fourier aplica la transformada y la multiplicacion con el filtro por canal dada una imagen, un filtro y una variable Inv para la inversa gausiana. Esta funcion es la encargada de realizar todo el trabajo de aplicar la tranformada de fourier al filtro gausiano, extrear los canales de la imagen pasarlos al dominio de la frecuencia y aplicar el filtro previamente creado, por ultimo simplemente multiplaca la imagen resultante con el filtro. Al final de todo el procedimiento nos va mostrando que va haciendo por canal mediante los plots.

```python
def gaussian_fft(sigma=0.25, size=512):
    [xx,yy]=np.meshgrid(np.linspace(-4,4,size),np.linspace(-4,4,size))
    g = np.exp(-0.5*(xx*xx+yy*yy)/(sigma*sigma))
    return g/np.max(g)

def fourier(image, gaussian_filter, inv = False, size = 512, hs = 55):
    
    image_crop = np.empty((image.shape[0],image.shape[1],image.shape[2]), dtype=float)
    filter_fft = fftpack.fftshift(gaussian_filter)
    if inv: filter_fft = 1.-fftpack.fftshift(gaussian_filter)
    channels = ['red','green','blue']
    
    hs = int(np.floor(hs/2.))
    fig, axarr = plt.subplots(ncols=5, nrows=1,figsize=(15,5))
    pos1 = axarr[0].imshow(fftpack.fftshift(np.log(1+np.abs(filter_fft))),cmap='jet')
    axarr[0].set_title('gaussian filter', size=16)
    
    for channel in range(0,3):
        image_fft = fftpack.fft2(image[:,:,channel], (size, size))
        res = image_fft * filter_fft
        
        pos = axarr[channel+1].imshow(fftpack.fftshift(np.log(1+np.abs(res))),cmap='jet')
        axarr[channel+1].set_title(channels[channel] + ' channel', size=16)
        fig.colorbar(pos, ax=axarr[channel+1])
        
        image_result = np.real(fftpack.ifft2(res))
        image_crop[:,:,channel]= image_result[hs:image.shape[0]+hs, hs:image.shape[1]+hs]
    
    
    copy = np.copy(image_crop)
    pos2 = axarr[4].imshow(normalize_img(copy))
    axarr[4].set_title('result', size=16)
    fig.colorbar(pos2, ax=axarr[4])
    plt.show()
    
    return image_crop


human_fft = fourier(img_as_float(human_crop), gaussian_filter=gaussian_fft(), hs=0)
cat_fft = fourier(img_as_float(cat), gaussian_filter=gaussian_fft(), inv=True, hs=0)

fig, axarr = plt.subplots(ncols=3, nrows=1,figsize=(15,5))
pos1 = axarr[0].imshow(human_fft)
pos2 = axarr[1].imshow(normalize_img(cat_fft))
pos3 = axarr[2].imshow(normalize_img(cat_fft + human_fft))
fig.colorbar(pos1, ax=axarr[0])
fig.colorbar(pos2, ax=axarr[1])
fig.colorbar(pos3, ax=axarr[2])
plt.show()    
```
